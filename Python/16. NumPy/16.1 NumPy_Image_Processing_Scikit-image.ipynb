# Get an image from skimage

```python
import numpy as np
```

```python
from skimage import data
```

```python
# loading image from skimage database

astronaut = data.astronaut()
```

```python
type(astronaut)
```

```python
astronaut.shape
```

```python
astronaut.size
```

```python
astronaut
```

# Plot the image data

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(astronaut)
```

# Cropping the image

```python
crop = astronaut[100:,50:,:]  # crops 100 points on x-axis, 50 on y-axis and third axis(RPG) remains unchanged

plt.imshow(crop)
```

# Spliting the image

```python
# Let's split it into two parts

a1,a2 = np.split(astronaut,2)
```

```python
# Display first part

plt.imshow(a1)
```

```python
# Display second part

plt.imshow(a2)
```

```python
# Let's split the image vertically

v1,v2 = np.split(astronaut,2, axis=1)
```

```python
plt.imshow(v1)
```

```python
plt.imshow(v2)
```

```python
# Alternatively, we can use vsplit and hsplit

v3,v4 = np.vsplit(astronaut,2)
plt.imshow(v3)
```

```python
# Alternatively, we can use vsplit and hsplit

h1,h2 = np.hsplit(astronaut,2)
plt.imshow(h1)
```

# Joining images

```python
# We can concatenate the image data

join = np.concatenate((a1,a2))
```

```python
plt.imshow(join)
```

```python
# Let's join horizontally

plt.imshow(np.concatenate((h1,h2), axis=1))
```

# Saving image

```python
plt.imsave("Astronaut.jpg",astronaut)
```
