## Tuples are similar to list except that...
 - Tuples are immutale:Once created, we can't change it within the program. e.g.: office codes,phone numbers, clothes sizes, etc.
 - Therefore, only uses read operations; never write operations. 
 - Therefore, we can't use operations like .append, .clear, .insert,...for tuples.
 - Therefore, iterating on a tuple is faster than list.
 - Use perenthesis ( ) for tuple and [ ] for list.

# Operations on tuples

### 5 basic operations that can be done on any sequence (lists, tuples, sets, dictionaries):
 1. Finding length: len()
 2. Concatenation
 3. membership
 4. repetition
 5. iteration

```python
# conatination

tuple3 = tuple1 + tuple2
print(tuple3)
```

```python
# repetition

semester_fees = ("Anna:2500", "Paul:3500")

program_fees = semester_fees * 3

program_fees
```

```python
# length of tuple3

len(tuple3)
```

```python
# membership

tuple4 = ("apples", "oranges", "pears", "peaches", "bananas")

"oranges" in tuple4
```

```python
"pears" not in tuple4
```

```python
26 in tuple3
```

```python
# we cannot append, extend, add to a tuple

tuple1.append("sugar")
```
