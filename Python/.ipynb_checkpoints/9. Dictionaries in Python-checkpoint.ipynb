## Dictionaries are a sequence such that...
 - each element is a Key:value pair.
 - like lists and tuples, the Key:Value pairs are separated by comas.
 - The Key:Value pairs are placed inside curly brackets {}. 
 - The Values in the dictionary are searched through their keys.
 - Indexing does not apply to dictionaries.
 - Keys must be unique: Duplicate keys not allowed
 - Keys are immutable: Tuples can be keys but not lists

```python
colors={10: "Red", 35: "Green", 15: "Blue", 25: "white"}

colors.items()
```

```python
# to sort mydict basesd on keys, use sorted(mydict.items(), key = lambda t: t[0])
colors={10: "Red", 35: "Green", 15: "Blue", 25: "white"}


sorted(colors.items(), key=lambda t: t[0])
```

```python
# to sort mydict basesd on values, use sorted(dict.items(), key = lambda t: t[1])

colors={10: "Red", 35: "Green", 15: "Blue", 25: "white"}


sorted(colors.items(), key=lambda t: t[1])
```

# Create a dictionary with input from user

```python
# Dictionary of employee salaries

n = int(input("Number of employees: ")) # user enters number of entries
salary_dict = {}
for item in range(n):
    name = input("Name: ")
    salary = input("Salary:")
    salary_dict.update({name: int(salary)})
print("The dictionary of employee salaries is:", salary_dict)
```

```python
salary_dict
```

```python
salary_dict["John"]
```

```python
for item in salary_dict:
    print(item)
```

## Sorting a dictionary using lambda function

```python
salary_dict
```

```python
# sort in alphabetical order of names

sorted(salary_dict.items())
```

```python
# sort in increasing salary amount

sorted(salary_dict.items(), key=lambda t:t[1])
```

# Zipping two lists into a dictionary

```python
country = ["Japan", "USA", "India", "China", "South Africa"]
capital = ["Tokyo", "DC", "New Delhi", "Beijing", "Cape Town"]

country_capital = list(zip(country, capital))

country_capital
```

```python
country = ["Japan", "USA", "India", "China", "South Africa"]
capital = ["Tokyo", "DC", "New Delhi", "Beijing", "Cape Town"]

country_capital = dict(zip(country, capital))

country_capital
```

# Convert a string to dictionary

```python
# firt, split the string into key-value pairs
string = "Joe-45,Popo-89,Mory-56,Ken-56"

kv_pair = string.split(",")

kv_pair
```

```python
# then create the key:value pair using for loop

lst=[]
for item in kv_pair:
    y=item.split("-")
    lst.append(y)

age_dict=dict(lst)

age_dict
    
    
    
```

```python
# finally, convert the ages to integers

age_dict1={}
for k,v in age_dict.items():
    age_dict1[k]=int(v)

print(age_dict1)
```
