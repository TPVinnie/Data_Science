# Programming languages
>### 1. Low level Languages: Closer to machine language. e.g.: Machine code, Assembly code.
>### 2. High Level languages: Closer to user language: e.g.: C, C++, Java, Python.

# Errors in Python
>### 1. Syntax errors: 
>>>Errors caught at compilation stage, grammar errors. e.g.: none use of ), undefined 
>### 2. Runtime errors: 
>>>Errors caught at interpretation stage, program began running but got into the error. e.g.: division by zero, undefined variable.
>### 3. Logic errors: 
>>>Errors in correct output due to incorrectly written program. Program is executed correctly without error but gives incorrect result. e.g.: error in mathematical formula

# Compiled Languages vs Interpreted Languages
> ## Compiler translates the whole code from one form to another form all at once.
> ## Interpreter translates instruction by instruction.
