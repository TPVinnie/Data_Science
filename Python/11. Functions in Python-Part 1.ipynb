# Function...

 - information is passed into a function as arguments.
 - A function may or may not return data.
 - In Python, a function is defined using the keyword def
 
 #### NB: Only one "return" can be used in one function
   ## Parameters vs. Arguments
  - Parameter are the variables placed inside parentheses in the function definition
  - Arguments are the values assigned for parameters when the function is called.
  
   ## Three types of functions in Python
 - **Built-in functions:** Predefined functions ready for use. e.g.: max(),min(), sorted()
 - **User-defined functions:** Created by us using the keyword def
 - **Anonymous function (aka lambda function):** Functioon without a name. Not declared with the def keyword.

## Default arguments

```python
# In the definition of a function, a default value can be assign for one of more parameters.
# When the function is called and no new argument is given, the default value is returned.

def veggies(veggy,price=50):
    '''the default price for most is 50. So, if we do not give a fdifferent price, 50 will be auto assigned''' 
    print(f"{veggy}:", price)
```

```python
veggies(veggy="Green pepper") # Here, the price is not given. The default price will be assigned
```

```python
veggies(price=67,veggy="Onion")
```

```python
# *args
# Function prints all elements passed to it

def PrintAll(*print_tuple):
    '''Give a list of coma-separated values and it will print the length and the items
    '''
    print("Number of items: ",len(print_tuple))
    print("List of items: ",print_tuple, end=",")
```

```python
PrintAll("apple","banana","orange","pear")
```

```python
PrintAll(3,7,7,4,3,9,965,6,1,1,0,8,6)
```

```python
# How can we add multiple indeterminate values?

# sum_integers_list.py
def my_sum(*args):
    result = 0
    for x in args:
        result += x
    return result

my_sum(1,2,3)
```

```python
# Function that sums up ant number of arguments passed to it

def mysum(*args):
    '''Takes two or more arguments and returns their sum.
    '''
    sums = 0
    for value in args:
        sums += value
    return sums

mysum(2,4,6,8,10)
```

```python
mysum(765,245,234,879,66)
```

## kwargs

```python
# Let's print some user details

def user_details(*args):
    print(args)

    print("We don't know what the above user details are for.To better understand what they are, we can assign keywords to them.\n Thus kwargs. Unlike args which is a tuple, this becomes a dictionary.")
```

```python
user_details('Powoh',43,85.5,'black','cameroon',49)
```

```python
# kwargs

def userdetails2(**kwargs):
    '''Takes keywords assigned to values as in keyword="Value". As many as are available.
    Prints the keyword-value sequence as a dictionary.
    '''
    return kwargs
```

```python
print(userdetails2(Name="Powoh", Age=43, Height=185.5,
             Eyes="black", Nationality="Cameroon", Size=49))
```

```python
# **kwargs
# Function prints a dictionary of capital cities; any given number.

def Capital(**country-city):
    '''Function prints a dictionary of capital cities; any given number.
    '''
    for key in country-city:
        print(key + ":" + country-city[key])
        
Capital(India="New Delhi",Canada="Ontario",Japan = "Tokyo")
```

# Local and Global variable
 - **Local variable:** Variable declared inside the function
 - **Gobal Variable:** Variable declared outside the function

```python
a=7
def mult(x):
    b=2
    return (a-b)*x

print("a= ",a)         #a is a global variable.
#print("b= ",b)        # b is a local variable in mult(). So it can't be printed outsid mult()
print("mult(5) = ",mult(5))
```

```python
# inside the function the local variable takes precedence. Outside, the global variable takes precedence.

a=1
def myFunction():
    a=2
    print("a= ",a)    #this prints the local a  
myFunction()
print("a*= ",a)       # this prints the glovbal a
```

# Prime numbers

```python
# this program prints all prime numbers in a given range

LB = int(input("Enter Lower Bound: "))
UB = int(input("Enter Upper Bound: "))


print("The prime numbers between {} and {} are: ".format(LB,UB))
for i in range(LB,UB + 1):
    for j in range(2,i):
        if (i%j) == 0:
            break
    else:
        print(i)
        
```
