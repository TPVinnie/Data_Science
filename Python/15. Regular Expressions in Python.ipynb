# Regular Expressions (RegEx) 

 - Used to search, match, find and split information  from data as per our requirements.
 - Available in all languages
 - Available methods in regex include: compile(), search(), match(), findall(), split(), etc.
 - Before using regex, the re module should be imported

# Sequence characters in RegEx
 - Sequence characters match only one character in the string

```python
reg = r'c\w\w' 

# this finds a string of three characters starting with c and followed by any 2 alphanumeric characters
```

```python
import re

mystr = ' cat mat rat hat bat can cap cp9 c&5'
reg = r'c\w\w'
# find all occurences of the above pattern
print(re.findall(reg,mystr))   # returns a list
```

# 3 Methods to remember are search(), findall(), and sub()
 - re.search("string",txt) returns the span (start and end position) of the first ocurence of "string" in "txt".
 - findall("string",txt) returns all the ocurences of "string" in "txt".
 - sub("string","strings",txt) replaces (substitutes) "string" with "strings" in "txt".

```python
# search - returns a match object. So, to get the output we need to apply group()
# match - returns a match object.  So, to get the output we need to apply group()
# findall - returns a list object.
# split - returns a list object.
# sub - returns a string object.
```

## match method
 - Only returns object if found in thew beginning, otherwise it returns "none"

```python
import re

mystr = 'cat mat rat hat bat can cap cp9 c&5'
reg = r'c\w\w'
# find all occurences of the above pattern
print(re.match(reg,mystr))   # returns a match object
```

```python
# apply the group() method to get the match

import re

mystr = 'cat mat rat hat bat can cap cp9 c&5'
reg = r'c\w\w'
# find all occurences of the above pattern
print(re.match(reg,mystr).group())   # return the first string
```

```python
reg1 = r'b\w\w'
match = re.match(reg1,mystr)        # return "none" because c doesn't occur at the beginning of mystr
print(match)  
```

## search method
 - returns the match object of the first occurence in the string

```python
reg1 = r'b\w\w'
search = re.search(reg1,mystr)       
print(search) 
```

```python
reg1 = r'b\w\w'
search = re.search(reg1,mystr)       
print(search.group()) 
```

```python
# We can get the span

reg1 = r'b\w\w'
search = re.search(reg1,mystr)       
print(search.span()) 
```

## findall method
 - returns all the occurences in a list object

```python
mystr1 = 'cat mat rat hat bat can cap cp9 b&5 lab pat gap cip bit pit'
reg1 = r'b\w\w'
found = re.findall(reg1,mystr1)       
print(found)
```

## split method
 - splits the string at into groups at the endpoints of the regex.

```python
mystr1 = 'cat mat rat hat bat can cap cp9 b&5 lab pat gap cip bit pit'
reg1 = r'b\w\w'
split = re.split(reg1,mystr1)       
print(split) 
```

## sub method
 - substitutes the regex with the provided string
 - sub(reg, "new_string", mystr) will substitute all the regex(s) found in mystr with new_string. 

```python
mystr1 = 'cat mat rat hat bat can cap cp9 b&5 lab pat gap cip bit pit'
reg1 = r'b\w\w'
substitute = re.sub(reg1, "BOT", mystr1)       
print(substitute) 
```

# Special characters in RegEx
 - They add more precision to the matching

```python
#Check if the string starts with "Hi"
import re
txt = "Hi. Good luck with your studies. Good luck with your exams. Happy everyday!"

y = re.findall("^Hi", txt)

print(y)

if y:
  print("Yes, there is a match!")
else:
  print("No match")
```

```python
#Check if the string ends with "everyday":

x = re.findall("everyday$", txt)

print(x)

if x:
  print("Yes, there is a match!")
else:
  print("No match")
```

```python
# Check if the string ends with "day!":

z = re.findall("day!$", txt)

print(z)

if z:
  print("Yes, there is a match!")
else:
  print("No match")
```
