# Three types of errors:

 

 - **Compile-Time Errors:** Caught by compiler. They are Syntax errors, Indentation errors, EOF/EOL errors.
 
 
 - **Runtime Errors:** Caught by PVM/Interpreter. They include: NameError, TypeError, FileNotFoundError, OverFlowError, ZeroDivisionError, etc.
 
 
 - **Logical Errors:** Due to wrong logic. Program runs successfully but the result is not as expected. For instance, the programmer writes the wrong equation.

**Exception handling takes care of runtime errors, only**

**Those errors which can be handled using exception handling methods are called exceptions**

Therefore, exceptions are a subset of errors: **All exceptions are errors but not all errors are exceptions**

## Exception is handled using try... except...else...finally
The programmer may only use try..except, try..except...else or all four. Depends on them

 - **try:** Used only for the part that an error is expected
 
 
 - **except:** write the code that will be executed if there is an exception
 
 
 - **Else:** Write the code that should be executed if there is no exception
 
 
 - **finally:** Write the clean up code. Maybe write code that deletes references to the inputs; maybe closing of file that raise the error while opening.

```python
print("Hey")
print("What's your name?")
try:
    print(Techo)                          # this is the suspiscious code. Likely to cause error/esception
except:
    print("Error: Enter name in quotes")   # this code will be executed if there is an exception in the try code
    
print("This is Python programming")
print("Thank you, Bye")
```

# How it works

```python
# No exception is caught


try:               # this will execute normally if there is no exception
    a = 10/0
    print(a)       

except:                             # this will execute if there is an exception
    print("error with the division")
    
    
else:                      # Optional block: this will execute if there is no exception. Doesn't execute if there is an exception
    print("No exception")
    
finally:             # this will always execute, with or witthout exception
    print("We must continue working...")
    print("Done!")
```

```python
# There is an exception

try:               # this will execute normally if there is no exception
    a = 10/0
    print(a)       

except:                             # this will execute if there is an exception
    print("error with the division")
    
    
else:                     # Optional block: this will execute if there is no exception. Doesn't execute if there is an exception
    print("No exception")
    
finally:             # this will always execute, with or witthout exception
    print("We must continue working...")
    print("Done!")
```

```python
# We can have more than one except block

try:
    a = 10/3
    b = "one"
    print(a) 
    print(a +b)

except ZeroDivisionError:                             
    print("Can't divide by zero")
    
    
except TypeError:
    print("Can't concatinate string with float")
    
    
else:                     
    print("No exception")
    
finally:             
    print("We must continue working...")
    print("Done!")
```

# Finding out where trhe exception occurred
 - We import the traceback module
 Then use the method traceback.print_exc()

```python
import traceback

try:
    a = int(input("Numerator: "))
    b = int(input("Denominator:"))
    c = a/b

except Exception as e:                             
    print("Error!: ",e)
    print(traceback.print_exc())
    
else:                     
    print("Numerator/Denominator = ",c)
    
finally:             
    print("We must continue working...")
```

# Raising exceptions
 - we use "raise" to raise an exception

```python
# we can raise an exception based on some condition
# Here,we are looking for square roots of numbers. If the user inputs a negative number, an exception is raised

try:
    number = int(input("Enter a positive number: "))
    if number < 0:
        raise Exception(number)
except:
    print("The number you entered is negative")
else:
    print(f"The square root of {number} is ", number**0.5)
finally:
    print("Done")
        
```

```python
# OverFlow Error: raised when the result of a calculation is too large

from math import exp
try:
    print(exp(1000))
except Exception as e:
    print(e)
    print(traceback.print_exc())
else:
    print ("Success!")
    
```

```python
# NameError: raised when a variable does not exist

try:
    print(x)
except Exception as e:
    print(e)
    print(traceback.print_exc())
else:
    print ("Success!")
```

```python
# AssertionError: raised when an assert statement is not found

try:
    a = 50
    b= "techo"
    assert a == b
except Exception as e:
    print(e)
    print(traceback.print_exc())
else:
    print ("Success!")
```

```python
# ModuleNotFoundError: raised when an imported module does not exist

try:
    import myModule
except Exception as e:
    print(e)
    print(traceback.print_exc())
else:
    print ("Success!")
```

# Assert statement
 - assert is used to check saome small conditions
 - It returns either "True" or "False"
 - An exception is raised when the condition is false

```python
help("assert")
```

```python
try:
    a = 50
    b = 50
    assert a == b
except Exception as e:
    print(e)
    print(traceback.print_exc())
else:
    print ("Success!")
```

```python
try:
    a = 50
    b = 200
    assert a == b
except AssertionError as e:
    print(e)
    print(traceback.print_exc())
else:
    print ("Success!")
```

```python
try:
    x = int(input("Enter a number between 50 and 100: "))
    assert x >= 50 and x <= 100
except AssertionError as e:
    print("Sorry, not in the correct range. Review and try again")
    print(traceback.print_exc())
else:
    print ("Great! The number you entered is in  the correct range")
```
