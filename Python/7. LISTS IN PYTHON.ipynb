## Syntax for list comprehension

```python
# The old way of printing the squares of the first 100 integers

my_list3 = []
for item in range(0, 101):
    my_list3.append(item ** 2)
print(my_list3)
```

```python
# Using list comprehension to print the squares of the first 100 integers

my_list4 = [item ** 2 for item in range(1, 101)]
print(my_list4)
```

```python
# Find the remainders when the first 100 squares are divided by 5

remainders = [x ** 2 % 5 for x in range(1, 101)]
print(remainders)
```

```python
# Find the remainders when the first 100 squares are divided by 11

remainders = [x ** 2 % 11 for x in range(1, 101)]
print(remainders)
```

```python
# Given the list of movies below

movies = ["Star Wars", "Ghandi", "Casablanca", "Shawshank Redemption", "Toy Story", "Raiders of the Lost Arc",
          "Ghostbusters", "Goodwill Hunting", "A space Odessey", "Coach Carter", "Welcome to America",
          "Game of Thrones"]
print(movies)
```

```python
# Find the movies that start with the letter G without list comprehension

moviesG = []
for item in movies:
    if item.startswith("G"):
        moviesG.append(item)
print(moviesG)
```

```python
# With list comprehension

Gmovies = [item for item in movies if item.startswith("G")]
print(Gmovies)
```

```python
# Find the movies released before 2000

moviesY = [("Star Wars", 1994), ("Ghandi", 2006), ("Casablanca", 1958), ("Shawshank Redemption", 2001), ("Toy Story", 2006), ("Raiders of the Lost Arc", 2003),
           ("Ghostbusters", 1999), ("Goodwill Hunting", 2011), ("A space Odessey",
                                                                1989), ("Coach Carter", 2007), ("Welcome to America", 1994),
           ("Game of Thrones", 1998)]
print(moviesY)
```

```python
# Find the movies released before 2000

movies2k = [title for (title, year) in moviesY if year < 2000]

print(movies2k)
```

```python
# Find the catesian product of two sets
# if A = {a,b} and B = {x,y}, Cartesian product = {(a,x), (a,y), (b,x), (b,y)}

A = [1, 3, 5, 7]
B = [2, 4, 6, 8]
cart = [(a, b) for a in A for b in B]
print(cart)
```
