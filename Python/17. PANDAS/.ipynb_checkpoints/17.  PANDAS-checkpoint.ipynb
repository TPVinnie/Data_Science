# Series
A one-dimensional array-like sequence of similar data types (like Numpy types) and associated indexes.

# Creating Series from lists (3 ways)
 - Series is created using the method .Series()
 - .Series(list,labels) creates a Series with values from "list" and their corresponding labels from "labels". If no labels are passed, pd creates labels with Python indexes.

### 1 Pass a list into the Series object

```python
# first import pandas

import pandas as pd # alternatively, you can use " from pandas import Series, DataFrame"


# then create a Series from a list

ser1 = pd.Series([-6,5,-3,5,7])
ser1
```

```python
# we can check the type of ser1

type(ser1)
```

```python
# we can get the values on the ser1

ser1.values
```

```python
# we can get the index object of ser1

ser1.index
```

```python
# check for the presence of a value

5 in ser1
```

#### Assigning custom indexes

```python
# It would be more expressive to assign our own indexes to the values

ser2 = pd.Series([6,4,-8,4,-2,4,1,0], index=['a','b','c','d','e','f','g','h'])

ser2
```

```python
ser2.values
```

```python
ser2.index
```

```python
 # we can check for the presence of a label

'b' in ser2
```

```python
'h' in ser2
```

```python
'i' in ser2
```

### 2. Pass a list and it's labels (labels become custom indexes)

```python
# we can assign values and labels


my_list = [5,7,-34,0,-8,6,90]

labels = ['a','b','c','d','e','f','g']


my_series = pd.Series(data=my_list,index=labels)

my_series
```

### 3. Simply pass the data and their labels

```python
# we can simply pass the data an the labels

my_list1 = [3,5,-29,-5,-13,1,85]
labels1 = ['a','b','c','d','e','f','g']

my_series1 = pd.Series(my_list1,labels1)
my_series1
```

```python
type(my_series)
```

```python
# we can get the values on the ser1

my_series1.values
```

```python
# we can get the labels of my+series1

my_series1.index
```

```python
# Let's display top 10 most populated countries (in millions, rounded to the nearest 10th million) as of 2022

pop10_dict = {'China':1440, 'India':1380,'USA':330, 'Indonesia':270, 'Pakistan':220, 
            'Brazi':210, 'Nigeria':210, 'Bangladesh':160, 'Russia':150, 'Mexico':130}

import pandas as pd
pop10 = pd.Series(pop10_dict)
pop10
```

Statistics from [worldmeters.info](https://www.worldometers.info/world-population/population-by-country/)

```python
# let's use a custom sorting order for the Series
# we'll pass the dictionary and then a sorted list as the indexes.

countries = ['Brazi', 'China', 'Cameroon', 'India', 'Indonesia', 'Kenya', 'Nigeria', 'Pakistan',  
             'Mexico', 'Bangladesh']

pop10_2 = pd.Series(pop10_dict, index = countries)

pop10_2.name = "Population"
pop10_2.index.name = "Top 10 Countries"
pop10_2                                
```

  - "NaN" indicates that the labels are not in the dict, thus no values are not available (NA).
  - As Rusia and USA were not in our custom labels, they are left out of the Series

```python
# isnull and notnull are also methods

pop10_2.isnull()
```

```python
# isnull and notnull are also methods

pop10_2.notnull()
```

```python
# we can find the total population of the top 10 countries

cum_pop10 = pd.Series.cumsum(pop10)
cum_pop10.name = "Cumulative Population"
cum_pop10.index.name = "Top 10 Countries"

cum_pop10
```

# loc[ ] and iloc[ ]
 - Both aee important for selecting rows
 - loc is a name-based selector--argument is row name. loc includes the end value
 - iloc is an index-base selector--argument is index. iloc excludes the end index
 - **loc is not commonly used in industry since with large files we can't remember names. But we can find index from the DataFrame; thus, iloc is more useful**
 - see [this video](https://youtu.be/lJDtzZsmF0g)

```python
# We can use loc to see the info of the first customer


pay_df.loc["NATHANIEL FORD"]
```

```python
# We can use iloc to see the info of the second customer

pay_df.iloc[1]
```

```python
# We can use iloc to see the info of a range of customers

pay_df.iloc[5:16]
```

```python
# We can use loc to find the inrersection of a row and column

pay_df.loc["ALSON LEE", "TotalPay"]
```

```python
# We can use iloc to find the inrersection of a row and column

pay_df.iloc[5,3]
```

```python
# We can use loc to display a range of rows and columns

pay_df.loc["ALSON LEE":"ARTHUR KENNEY", "OvertimePay":"TotalPay"]
```

**Note how loc includes the end names**

```python
# We can use iloc to display a range of rows and columns

pay_df.iloc[6:10, 1:3]
```

**Note how iloc does not include the end values**

```python
# We can use loc to check for the present of a value in a specified column

pay_df.loc[pay_df["OvertimePay"] == 0.00]
```

```python
# Let's use loc to search the name "Roy I Tillery"

salaries_df.loc[salaries_df["EmployeeName"] == "Roy I Tillery"]
```

```python
# Let's use loc to search for all Custodians

custodianSalaries_df = salaries_df.loc[salaries_df["JobTitle"] == "Custodian"]
custodianSalaries_df
```
