```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
```

```python
webpage = requests.get('https://www.ambitionbox.com/list-of-companies?page=1',headers=headers).text
```

```python
webpage
```

```python
soup = BeautifulSoup(webpage, 'lxml')
soup
```

```python
soup.find_all('h1')[0].text
```

```python
soup.find_all('h2')
```

```python
for i in soup.find_all('h2'):
    print(i.text.strip())
```

```python
for i in soup.find_all('p'):
    print(i.text.strip())
```

```python
company = soup.find_all('div', class_='company-content-wrapper')
company
```

```python
c_name = []
rating = []
reviews = []
ctype = []
headquarter = []
how_old = []
no_of_employees = []

for i in company:
    c_name.append(i.find('h2').text.strip())
    rating.append(i.find('p', class_='rating').text.strip())
    reviews.append(i.find('a', class_='review-count').text.strip())
    ctype.append(i.find_all('p', class_='infoEntity')[0].text.strip())
    headquarter.append(i.find_all('p', class_='infoEntity')[1].text.strip())
    how_old.append(i.find_all('p', class_='infoEntity')[2].text.strip())
    no_of_employees.append(i.find_all('p', class_='infoEntity')[3].text.strip())
    
    
df = pd.DataFrame({'Comp_Name':c_name, "Rating" : rating, "Reviews":reviews,
                  "Comp_Type": ctype, "Comp_Head_Quarter": headquarter,
                  "Comp_Age" : how_old, "No. of Employes":no_of_employees})
```
